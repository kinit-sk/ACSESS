# Generate Grid Image (3x3)

```python
import os
import cv2
import matplotlib.pyplot as plt
from matplotlib import gridspec
```

```python
dataset_name = 'insects'
dataset_sample_images_directory = './insects_samples'
```

```python
ds_imgs = os.listdir(dataset_sample_images_directory)

if ".ipynb_checkpoints" in ds_imgs:
    i = ds_imgs.index(".ipynb_checkpoints")
    del ds_imgs[i]

ds_imgs_with_dir = [os.path.join(dataset_sample_images_directory,imm) for imm in ds_imgs]

nrow = 3
ncol = 3

fig = plt.figure(figsize=(ncol+1, nrow+1)) 

gs = gridspec.GridSpec(nrow, ncol,
     wspace=0.04, hspace=0.04, 
     top=1.-0.5/(nrow+1), bottom=0.5/(nrow+1), 
     left=0.5/(ncol+1), right=1-0.5/(ncol+1)) 


img_index = 0
for i in range(3):
    for j in range(3):
        img = cv2.imread(ds_imgs_with_dir[img_index])
        img_index += 1
        im_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        ax= plt.subplot(gs[i,j])
        ax.imshow(im_rgb)
        ax.set_xticklabels([])
        ax.set_yticklabels([])
        ax.axis('off')
plt.show()
saveTo = os.path.join(dataset_sample_images_directory,dataset_name+'_grid.png')
fig.savefig(saveTo,bbox_inches='tight', dpi=fig.dpi)
```
